### 6. Limpeza dos Dados
Remove colunas desnecessárias do dataframe para simplificar a análise. Exibe as primeiras linhas do dataframe limpo.

```python
df_cleaned = df_rent.drop(["New", "Property Type", "Negotiation Type"], axis=1)
```

### 7. Codificação de Variáveis Categóricas
Utiliza `OrdinalEncoder` e `OneHotEncoder` para transformar a coluna de distritos em representações numéricas. Junta as colunas codificadas ao dataframe original, removendo a coluna categórica original.

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

district_encoded = ordinal_encoder.fit_transform(df_cleaned[["District"]])
district_encoded
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()

housing_cat_1hot = cat_encoder.fit_transform(df_cleaned[["District"]])
housing_cat_1hot
```

```python
one_hot = pd.get_dummies(df_cleaned["District"])

df = df_cleaned.drop('District',axis = 1)
df = df.join(one_hot)
df
```

### 8. Divisão dos Dados em Conjuntos de Treinamento e Teste
Separa os dados em variáveis dependentes (`Y`) e independentes (`X`). Divide os dados em conjuntos de treinamento e teste.

```python
from sklearn.model_selection import train_test_split
```

```python
Y = df["Price"]
X = df.loc[:, df.columns != "Price"]
```

```python
X.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

### 9.1. Treinamento e Avaliação com Regressão Linear
Treina um modelo de regressão linear com os dados de treinamento. Faz previsões para algumas amostras e compara com os rótulos reais. Calcula o erro quadrático médio (MSE) e a raiz do erro quadrático médio (RMSE) para avaliar o modelo

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python
some_data = X.iloc[:5]
some_labels = Y.iloc[:5]

print("Predictions:", lin_reg.predict(some_data))
print("Labels:", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error

preds = lin_reg.predict(x_train)
lin_mse = mean_squared_error(y_train, preds)

lin_rmse = np.sqrt(lin_mse)
lin_rmse
```


### 9.2. Treinamento e Avaliação com Regressor de Árvore de Decisão
Treina um modelo de árvore de decisão com os dados de treinamento. Faz previsões e calcula o MSE e RMSE para avaliar a performance do modelo.

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(x_train, y_train)
```

```python
preds = tree_reg.predict(x_train)
tree_mse = mean_squared_error(y_train, preds)

tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

### 9.3. Validação Cruzada com Árvore de Decisão
Utiliza `cross_val_score` para realizar validação cruzada no modelo de árvore de decisão. Calcula e exibe as pontuações, a média e o desvio padrão dos resultados.

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, x_train, y_train,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
                           
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse_scores)
```

### 9.4. Validação Cruzada com Regressão Linear
Utiliza `cross_val_score` para realizar validação cruzada no modelo de regressão linear. Calcula e exibe as pontuações, a média e o desvio padrão dos resultados.

```python
scores = cross_val_score(lin_reg, x_train, y_train,
                         scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-scores)
                           
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(lin_rmse_scores)
```

### 10. Treinamento com Regressor de Floresta Aleatória
Treina um modelo de regressão de floresta aleatória (`RandomForestRegressor`) com os dados de treinamento.

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor()
rf_reg.fit(x_train, y_train)
```

```python
preds = rf_reg.predict(x_train)
rf_mse = mean_squared_error(y_train, preds)

rf_rmse = np.sqrt(rf_mse)
rf_rmse
```

### 11. Avaliação do Modelo de Floresta Aleatória
Faz previsões utilizando o modelo de floresta aleatória e calcula o erro quadrático médio (MSE) e a raiz do erro quadrático médio (RMSE) para avaliar o modelo. Realiza validação cruzada e exibe as pontuações, a média e o desvio padrão dos resultados.

```python
scores = cross_val_score(rf_reg, x_train, y_train,
                         scoring="neg_mean_squared_error", cv=10)
rf_rmse_scores = np.sqrt(-scores)
                           
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(rf_rmse_scores)
```

### 12. Otimização de Hiperparâmetros com Grid Search
Utiliza `GridSearchCV` para realizar a busca de hiperparâmetros no modelo de floresta aleatória. Define uma grade de parâmetros (`param_grid`) para testar várias combinações de hiperparâmetros. Realiza a busca e treina o modelo com os dados de treinamento.

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv=5, 
                           scoring='neg_mean_squared_error',
                           return_train_score=True)

grid_search.fit(x_train, y_train)
```

### 13. Avaliação dos Resultados do Grid Search
Exibe os melhores parâmetros (`best_params_`) e o melhor estimador (`best_estimator_`) encontrados pelo `GridSearchCV`. Itera pelos resultados de validação cruzada e exibe a média das pontuações para cada combinação de parâmetros testada

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

### 14. Avaliação Final do Modelo Otimizado
Utiliza o melhor estimador encontrado pelo `GridSearchCV` para fazer previsões no conjunto de teste. Calcula o erro quadrático médio (MSE) e a raiz do erro quadrático médio (RMSE) para avaliar a performance final do modelo.

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
final_model = grid_search.best_estimator_
final_model_predictions = final_model.predict(x_test)

final_mse = mean_squared_error(y_test, final_model_predictions)
final_rmse = np.sqrt(final_mse)

print(final_rmse)
```
